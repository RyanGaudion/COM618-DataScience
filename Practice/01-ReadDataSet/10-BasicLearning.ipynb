```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import preprocessing
```

```python
df = pd.read_csv('../data/heart_disease_health_indicators_BRFSS2021.csv')

df.info()
```

```python
# Predict BMI from Health Living Stats using multi-linear regression
X_props = ["FruitsPerDay", "DarkGreensPerDay", "PotatoesPerDay", "OtherVegPerDay", "BadMentalHealthLast30Days", "BadPhysicalHealthLast30Days", "AgeGroup"]

X = df[X_props]
y = df[["BMI"]]
```

```python
sns.scatterplot(x="FruitsPerDay", y="BMI", data=df)
```

```python
sns.scatterplot(x="BadPhysicalHealthLast30Days", y="BMI", data=df)
```

```python
sns.scatterplot(x="AgeGroup", y="BMI", data=df)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
print("MSE: ", mean_squared_error(y_test, predictions))
print("MAE: ", mean_absolute_error(y_test, predictions))
```

```python
predictedForMe = model.predict([[2, 1, 1, 2, 0, 0, 1]])
```

```python
print("Predicted BMI: ", predictedForMe)
```
