```python
import matplotlib.dates as mdates
df2['date'] = mdates.date2num(df2['date'])

# Plot dataframe with date as x-axis and y as y-axis
plt.plot(df2['date'], df2['Y'])

# Set range of date from 1860 to 1888
plt.xlim('1860', '1888')

# Plot horizontal green solid marker at Y=4
plt.axhline(y=4, color='green')

# Plot vertical red dash marker at X=1875
#plt.axvline(x='1875', color='red', linestyle='--')

# Plot red highlighted area between 1885 and 1890
#plt.axvspan('1885', '1890', color='red', alpha=0.5)

# Display Time series as title
plt.title('Time series')

# Display Number of discoveries on Y axis
plt.ylabel('Number of discoveries')

# Display Date on X axis
plt.xlabel('Date')

# Show plot
plt.show()
```
